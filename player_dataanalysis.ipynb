```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x="Season",data=df_matches,order=df_matches["Season"].value_counts().index.sort_values(),palette="muted")
plt.xlabel("Season",fontsize=12)
plt.ylabel("No. of matches",fontsize=12)
plt.title("No. of matches played in each ipl season",fontsize=12)
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x="Season",hue="toss_decision",data=df_matches,order=df_matches["Season"].value_counts().index.sort_values(),palette="deep")
plt.xlabel("Season",fontsize=12)
plt.ylabel("No. of matches",fontsize=12)
plt.title("No. of matches played in each ipl season",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(11,9))
sns.countplot(y="winner",data=df_matches,order=df_matches["winner"].value_counts().index,palette="copper")
plt.xlabel("no. of wins",fontsize=12)
plt.ylabel("teams",fontsize=12)
plt.title("No. of matches won by the teams",fontsize=12)
```

```python
plt.figure(figsize=(11,9))
sns.countplot(y="venue",data=df_matches,order=df_matches["venue"].value_counts().iloc[:10].index,palette="gist_heat")
plt.xlabel("no. of matches",fontsize=12)
plt.ylabel("venues",fontsize=12)
plt.title("No. of matches played in each stadium",fontsize=12)
```

```python
plt.figure(figsize=(6,6))
sizes=df_matches.toss_decision.value_counts()
labels=df_matches.toss_decision.value_counts().index
plt.pie(sizes,colors = ['#ff9999','#66b3ff'],labels=labels,autopct="%1.1f%%",startangle=90,pctdistance=0.75,
         explode = (0.025,0.025))
center_circle=plt.Circle((0,0),0.6,fc="white")
fig=plt.gcf()
fig.gca().add_artist(center_circle)
```

```python
plt.figure(figsize=(15,7))
player_match=df_matches["player_of_match"].value_counts()[:10]
sns.barplot(player_match.index,player_match.values,palette="cividis")
plt.xlabel("player",fontsize=12)
plt.ylabel("count",fontsize=12)
plt.title("player of the match",fontsize=12)
```
